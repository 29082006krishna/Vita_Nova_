```python
# ============================
# Product-Specific Sales & Demand Forecasting
# Using your dataset column names
# ============================
!pip install openai
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from datetime import datetime
import calendar
from openai import OpenAI
```

```python

# -----------------------------
# Step 1: Load Data
# -----------------------------


df = pd.read_csv("dairy_dataset.csv")  # replace with your CSV path
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
# -----------------------------
# Step 2: User Input: Product & Forecast Period
# -----------------------------
product_name = input("Enter Product Name: ").strip()
state_name = input("Enter State: ").strip()
months_ahead = int(input("Enter number of months to forecast (3, 6, 12): "))

# Filter for product + state
df_product = df[
    (df['Product Name'] == product_name) &
    (df['Location'] == state_name)
].copy()

if df_product.empty:
    raise ValueError(
        f"No data found for product '{product_name}' in state '{state_name}'"
    )
# -----------------------------
# Step 3: Preprocess Data
# -----------------------------
# Select sales and demand columns
df_product_model = df_product[['Quantity Sold (liters/kg)', 'Reorder Quantity (liters/kg)']].copy()

# Scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_product_model.values)

# Create sequences
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data)-seq_length):
        x = data[i:(i+seq_length)]
        y = data[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

SEQ_LENGTH = 30
X, y = create_sequences(scaled_data, SEQ_LENGTH)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Convert to PyTorch tensors
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
y_train = torch.from_numpy(y_train).float()
y_test = torch.from_numpy(y_test).float()

train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# -----------------------------
# Step 4: LSTM Model
# -----------------------------
class LSTMModel(nn.Module):
    def __init__(self, input_size=2, hidden_size=50, num_layers=2, output_size=2):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

model = LSTMModel()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# -----------------------------
# Step 5: Train Model
# -----------------------------
EPOCHS = 50
for epoch in range(EPOCHS):
    model.train()
    for xb, yb in train_loader:
        optimizer.zero_grad()
        output = model(xb)
        loss = criterion(output, yb)
        loss.backward()
        optimizer.step()
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {loss.item():.6f}")

# -----------------------------
# Step 6: Predictions
# -----------------------------
model.eval()
y_actual = scaler.inverse_transform(y_test.numpy())
predicted_test = model(X_test).detach().numpy()
predicted_test = scaler.inverse_transform(predicted_test)



# -----------------------------
# Step 7: Graph 1 - Historical Sales & Demand
# -----------------------------
# plt.figure(figsize=(12,6))
# plt.plot(df_product['Date'].iloc[-len(y_test):], y_actual[:,0], label="Actual Sales")
# plt.plot(df_product['Date'].iloc[-len(y_test):], y_actual[:,1], label="Actual Demand")
# plt.title(f"Historical Sales & Demand for {product_name}")
# plt.xlabel("Date")
# plt.ylabel("Quantity (liters/kg)")
# plt.legend()
# plt.show()
df_last_12m = df_product.sort_values('Date').tail(12)
month_labels_hist = [f"{calendar.month_name[d.month]}-{d.year}" for d in df_last_12m['Date']]

plt.figure(figsize=(12,6))
plt.plot(month_labels_hist, df_last_12m['Quantity Sold (liters/kg)'], marker='o', label="Actual Sales")
plt.plot(month_labels_hist, df_last_12m['Reorder Quantity (liters/kg)'], marker='o', label="Actual Demand")
plt.title(f"Historical Sales & Demand (Last 12 Months) for {Product_name_input.title()}")
plt.xlabel("Month")
plt.ylabel("Quantity (liters/kg)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
# -----------------------------
# Step 6.5: Recursive Future Daily Prediction
# -----------------------------
model.eval()

# Predict required days (months  30)
days_ahead = months_ahead * 30
future_preds = []

# Start from last known sequence
last_sequence = X_test[-1].unsqueeze(0)  # (1, seq_len, 2)

for _ in range(days_ahead):
    with torch.no_grad():
        next_pred = model(last_sequence)  # (1, 2)
        future_preds.append(next_pred.squeeze(0).numpy())

        # Slide window forward
        next_pred_seq = next_pred.unsqueeze(1)  # (1, 1, 2)
        last_sequence = torch.cat(
            (last_sequence[:, 1:, :], next_pred_seq),
            dim=1
        )

future_preds = np.array(future_preds)

# Inverse scale
future_preds = scaler.inverse_transform(future_preds)

# -----------------------------
# Step 8: Graph 2 - Predicted Future Sales & Demand
# -----------------------------
# Get current month and year
import numpy as np

# Number of months to forecast
forecast_months = months_ahead

# Aggregate daily predictions into monthly predictions
daily_preds = np.array(future_preds)  # shape (days_ahead, 2)
monthly_preds = []

for i in range(forecast_months):
    start = i*30
    end = (i+1)*30
    monthly_avg = daily_preds[start:end].mean(axis=0)  # average sales & demand for the month
    monthly_preds.append(monthly_avg)

monthly_preds = np.array(monthly_preds)  # shape (months_ahead, 2)

# Generate month labels
from datetime import datetime
import calendar

now = datetime.now()
current_month = now.month
current_year = now.year

month_labels = []
for i in range(forecast_months):
    month_index = (current_month + i - 1) % 12 + 1
    year_offset = (current_month + i - 1) // 12
    month_labels.append(f"{calendar.month_name[month_index]}-{current_year + year_offset}")

# Plot predicted graph (monthly)
plt.figure(figsize=(12,6))
plt.plot(month_labels, monthly_preds[:,0], label="Predicted Sales", marker='o')
plt.plot(month_labels, monthly_preds[:,1], label="Predicted Demand", marker='o')
plt.title(f"Predicted Sales & Demand for next {forecast_months} months ({product_name})")
plt.xlabel("Month")
plt.ylabel("Quantity (liters/kg)")
plt.xticks(rotation=45)
plt.legend()
plt.show()


# Prepare numeric summary for AI
summary_data = {
    "product": product_name,
    "state": state_name,
    "forecast_months": months_ahead,
    "avg_sales": float(monthly_preds[:, 0].mean()),
    "avg_demand": float(monthly_preds[:, 1].mean()),
    "sales_trend": float(monthly_preds[-1, 0] - monthly_preds[0, 0]),
    "demand_trend": float(monthly_preds[-1, 1] - monthly_preds[0, 1])
}

# prompt = f"""
# You are an AI business analyst.

# Explain the demand forecasting results below in clear, professional language.

# Product: {summary_data['product']}
# State: {summary_data['state']}
# Forecast Period: {summary_data['forecast_months']} months
# Average Sales: {summary_data['avg_sales']:.2f}
# Average Demand: {summary_data['avg_demand']:.2f}
# Sales Trend Change: {summary_data['sales_trend']:.2f}
# Demand Trend Change: {summary_data['demand_trend']:.2f}

# Explain:
# 1. Sales and demand trends
# 2. Inventory risk (stock-out or overstock)
# 3. Business recommendation
# """

# response = client.chat.completions.create(
#     model="gpt-4o-mini",
#     messages=[
#         {"role": "system", "content": "You are an expert AI explainability assistant."},
#         {"role": "user", "content": prompt}
#     ],
#     temperature=0.3
# )

# ai_explanation = response.choices[0].message.content

# print("\n" + "="*60)
# print("AI GENERATED EXPLAINABLE SUMMARY (OpenAI)")
# print("="*60)
# print(ai_explanation)
# print("="*60)

```
